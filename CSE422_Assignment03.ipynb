#Problem 1: Chess Masters

```python
import math
import random

starter = int(input("Enter starting player for game 1 (0 for Carlsen, 1 for Caruana): "))
Carlsen = float(input("Enter base strength for Carlsen: "))
Caruana = float(input("Enter base strength for Caruana: "))

crlsn, crn, draw = 0, 0, 0

def strength(x):
    return (math.log2(x + 1) + (x / 10))

crlsnS = strength(Carlsen)
crnS = strength(Caruana)

def utility(maxV, minV):
    return (round((maxV - minV + (((-1) ** (random.choice([0, 1]))) * (random.randint(1, 10) / 10))), 2))

def chess_master(depth, alpha, beta, maxV, minV, maxP):
    if depth == 0:
       return utility(maxV, minV)
    if maxP == False:
       bestV = (math.inf)
       for i in range(2):
           V = chess_master(depth -1, alpha, beta, maxV, minV, True)
           bestV = min(bestV, V)
           beta = min(beta, V)
           if beta <= alpha:
               break
       return bestV
    else:
       bestV = (- math.inf)
       for j in range(2):
           V = chess_master(depth -1, alpha, beta, maxV, minV, False)
           bestV = max(bestV, V)
           alpha = max(alpha, V)
           if beta <= alpha:
               break
       return bestV

for k in range(1, 5):
    currentP = (starter + k - 1) % 2
    if currentP == 0:
        maxP, minP = "Magnus Carlsen (Max)", "Fabiano Caruana (Min)"
        maxV, minV = crlsnS, crnS
    else:
        maxP, minP = "Fabiano Caruana (Max)", "Magnus Carlsen (Min)"
        maxV, minV = crnS, crlsnS
    res = chess_master(5, - math.inf, math.inf, maxV, minV, currentP == 0)
    if res > 0:
        winner = maxP
        if "Magnus Carlsen" in winner:
            crlsn += 1
        else:
            crn +=1
    elif res < 0:
        winner = minP
        if "Magnus Carlsen" in winner:
            crlsn += 1
        else:
            crn +=1
    else:
        winner = "Draw"
        draw += 1
    print(f"Game {k} Winner: {winner} (Utility value: {res})")

if crlsn > crn:
    winner = "Magnus Carlsen"
elif crlsn < crn:
    winner = "Fabiano Caruana"
else:
    winner = "draw"
print("Overall Results:")
print(f"Magnus Carlsen Wins: {crlsn}")
print(f"Fabiano Caruana Wins: {crn}")
print(f"Draws: {draw}")
print(f"Overall Winner: {winner}")
```
