```python
import heapq

def Graph(input_file):
    graph = {}
    hrstc = {}
    file = open(input_file, "r")
    f = file.readlines()
    for i in f:
        text = i.split()
        node = text[0]
        hrstc[node] = int(text[1])
        edge =[]
        for j in range(2, len(text), 2):
            adj_node = text[j]
            dist = int(text[j + 1])
            edge.append((adj_node, dist))
        graph[node] = edge
    file.close()
    return graph, hrstc

def search(graph, hrstc, s, d):
    pq = []
    heapq.heappush(pq, (hrstc[s], 0, s, [s]))
    visited = []
    while len(pq) > 0:
        low_node = heapq.heappop(pq)
        _, cost, node, path = low_node
        if node == d:
            return path, cost
        if node not in visited:
            visited.append(node)
        for adj_node, dst in graph[node]:
            if adj_node not in visited:
                new_cost = cost + dst
                total_cost = new_cost + hrstc[adj_node]
                new = path + [adj_node]
                heapq.heappush(pq, (total_cost, new_cost, adj_node, new))
    return None, None

input_file = "input.txt"
graph, hrstc = Graph(input_file)
Start_node = input("Enter your starting node: ")
Destination = input("Enter your destination: ")
path, total_dst = search(graph, hrstc, Start_node, Destination)
if path:
    print(f"Path: {' -> '.join(path)}\nTotal distance: {total_dst} km")
else:
    print("NO PATH FOUND")
```
