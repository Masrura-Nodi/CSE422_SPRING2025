```python
import random

initial_capital = 1000
prices = [-1.2, 3.4, -0.8, 2.1, -2.5, 1.7, -0.3, 5.8, -1.1, 3.5]
population_size = 4
generations = 10
mutation_chance = 0.05

def initialize_population():
    population = []
    for i in range(population_size):
        population.append({"stop_loss": (random.randint(1, 99)), "take_profit": (random.randint(1, 99)), "trade_size": (random.randint(1, 99))})
    return population

def  evaluate_fitness(s):
    capital = initial_capital
    for j in prices:
        trade = capital * (s["trade_size"] / 100)
        if j <= -s["stop_loss"]:
            capital -= trade * (s["stop_loss"] / 100)
        elif j >= s["take_profit"]:
            capital += trade * (s["take_profit"] / 100)
        else:
            capital += trade * (j / 100)
    return capital - initial_capital

def selection(population):
    p1 = random.choice(population)
    p2 = random.choice(population)
    while p1 == p2:
        p2 = random.choice(population)
    return p1, p2

def crossover(p1, p2):
    cpoint = random.randint(0, 2)
    c1 = {}
    c2 = {}
    keys = list(p1.keys())
    for i in range(len(keys)):
        if i > cpoint:
            c1[keys[i]] = p2[keys[i]]
            c2[keys[i]] = p1[keys[i]]
        else:
            c1[keys[i]] = p1[keys[i]]
            c2[keys[i]] = p2[keys[i]]
    return c1, c2

def mutation(s):
    if random.random() < mutation_chance:
        mutate = random.choice(list(s.keys()))
        s[mutate] = random.randint(1, 99)
    return s

def new(population):
    best_strategy = None
    best_profit = float('-inf')
    for s in population:
        profit = evaluate_fitness(s)
        if profit > best_profit:
            best_profit = round((profit),2)
            best_strategy = s
    return best_strategy, best_profit

def output():
    population = initialize_population()
    best_strategy, best_profit = new(population)
    for k in range(generations):
        newp = []
        while len(newp) < population_size:
            p1, p2 = selection(population)
            c1, c2 = crossover(p1, p2)
            c1 = mutation(c1)
            c2 = mutation(c2)
            newp.append(c1)
            newp.append(c2)
        population = newp[:population_size]
        best_strategy, best_profit = new(population)
    op = f"best_strategy: {best_strategy} \nFinal_profit: {best_profit}"
    return op

result = output()
print(result)
```
